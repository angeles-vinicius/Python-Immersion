```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Python Immersion.xlsx", sheet_name="Principal")
df_principal.head(10)
```

```python
df_Total_de_acoes = pd.read_excel("/content/Python Immersion.xlsx", sheet_name="Total_de_acoes")
df_Total_de_acoes
```

```python
df_Ticker = pd.read_excel("/content/Python Immersion.xlsx", sheet_name="Ticker")
df_Ticker
```

df_Chatgpt = pd.read_excel("/content/Python Immersion.xlsx", sheet_name="Chatgpt")
df_Chatgpt

```python
df_principal = df_principal[["Ativo",	"Data",	"Último (R$)",	"Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Último (R$)":"valor_final", "Var. Dia (%)":"var_dia_pct"}).copy()
df_principal
```

```python
df_principal["var_pct"] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] =  df_principal['valor_final'] / (df_principal["var_pct"] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_Total_de_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal["variacao_rs"] = (df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal["Qtde. Teórica"]
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={"Qtde. Teórica":"qtd_teorica"}).copy()
df_principal
```

```python
df_principal['status'] = pd.Series(['up' if x > 0 else 'down' if x < 0 else 'no changes' for x in df_principal['variacao_rs']])
df_principal
```

```python
df_principal = df_principal.merge(df_Ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_Chatgpt, left_on='Nome', right_on='Nome da empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={"Idade (anos)":"idade"}).copy()
df_principal
```

```python
df_principal['cat_idade'] = df_principal['idade'].apply(lambda x: "estabelecida" if x > 100 else "madura" if x >= 50 else "em crescimento" if x >= 20 else "startup")
df_principal
```

```python
# Calcule o maior valor da coluna 'variação_rs'
maior = df_principal['variacao_rs'].max()

# Calcule o menor valor da coluna 'variação_rs'
menor = df_principal['variacao_rs'].min()

# Calcule a média dos valores da coluna 'variação_rs'
media = df_principal['variacao_rs'].mean()

# Filtre os valores da coluna 'variação_rs' onde o status é 'up' e calcule a média
media_subiu = df_principal[df_principal['status'] == 'up']['variacao_rs'].mean()

# Filtre os valores da coluna 'variação_rs' onde o status é 'down' e calcule a média
media_desceu = df_principal[df_principal['status'] == 'down']['variacao_rs'].mean()

# Imprima os resultados
print("Maior\t", "R$", "{:,.2f}".format(maior))
print("Menor\t", "R$", "{:,.2f}".format(menor))
print("Média\t", "R$", "{:,.2f}".format(media))
print("Méida Subiu\t", "R$", "{:,.2f}".format(media_subiu))
print("Média Desceu\t", "R$", "{:,.2f}".format(media_desceu))
```

```python
fig = px.bar(df_analise_saldo, x='status', y='variacao_rs', text='variacao_rs', title='Variação Reais por Resultado')
tickvals = df_analise_saldo['variacao_rs']
ticktext = ['R$ {:,.2f}'.format(val) for val in tickvals]
fig.update_yaxes(tickvals=tickvals, ticktext=ticktext)
fig.show()
```

```python
fig = px.pie(df_analise_segmento, values='variacao_rs', names='Segmento', title='Variação Reais por Segmento')
fig.show()
```
